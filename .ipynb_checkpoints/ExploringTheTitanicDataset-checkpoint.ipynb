```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv("datasets/titanic_train.csv")
titanic_df.head(10)
```

```python
titanic_df.shape
```

```python
titanic_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'],'columns', inplace = True)
titanic_df.head(10)
```

```python
#check for missing values
titanic_df[titanic_df.isnull().any(axis=1)].count()
```

```python
#remove records with NaN fields
titanic_df = titanic_df.dropna()
```

```python
titanic_df.shape
```

```python
#missing fields removed
titanic_df[titanic_df.isnull().any(axis=1)].count()
```

```python
titanic_df.describe()
```

```python
titanic_df.head()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(titanic_df['Age'], titanic_df['Survived'])
plt.xlabel('Age')
plt.ylabel('Survived')
```

```python
 fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(titanic_df['Fare'], titanic_df['Survived'])
plt.xlabel('Fare')
plt.ylabel('Survived')
```

```python
pd.crosstab(titanic_df['Sex'], titanic_df['Survived'])
```

```python
pd.crosstab(titanic_df['Pclass'], titanic_df['Survived'])
```

```python
#correlation matrix
titanic_data_corr = titanic_df.corr()
titanic_data_corr
```

```python
#visualize correlation data, Light colors indicate positive correlation, dark colors indicate negative correlation
fig, ax = plt.subplots(figsize=(12,10))
sns.heatmap(titanic_data_corr, annot=True)
```

```python
#preprocessing set Sex attribute values to numeric, 0 represents female, 1 represents male
from sklearn import preprocessing
label_encoding = preprocessing.LabelEncoder()
titanic_df['Sex'] = label_encoding.fit_transform(titanic_df['Sex'].astype(str))
titanic_df.head()
```

```python
label_encoding.classes_
```

```python
#categorical, non-binary values with no inherent order can be converted to numeric using one-hot encoding
titanic_df = pd.get_dummies(titanic_df, columns = ['Embarked'])
titanic_df.head()
```

```python
#shuffle records, drop original indices
#shuffling ensures the ML model doesnt pick up patterns in the Ordering of the data
titanic_df = titanic_df.sample(frac=1).reset_index(drop=True)
titanic_df.head()
```

```python
titanic_df.to_csv('datasets/titanic_processed.csv', index = False)
!ls datasets
```
